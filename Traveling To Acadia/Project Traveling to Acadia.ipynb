```python
import numpy as np
from matplotlib import pyplot as plt

#Loading the data:

#in_bloom is a NumPy array containing 900 numbers. Each number is a day of the year, which represents the average start date of a flower blooming:
in_bloom = np.loadtxt(open("in-bloom.csv"), delimiter=",")
#flights is a NumPy array with 11,000 numbers. Each number is a day of the year, which represents a flight from your hometown to airports near Acadia, Maine:
flights = np.loadtxt(open("flights.csv"), delimiter=",")

#Plottinh the histograms:

plt.figure(1)

plt.subplot(211)
plt.hist(flights, range = (0, 365), bins = 365, color = 'DarkSlateBlue')
plt.title("Flights by Day")
plt.xlabel("Day of the Year")
plt.ylabel("Flight Count")

plt.subplot(212)
plt.hist(in_bloom, range = (0, 365), bins = 365, color = 'DarkSlateBlue')
plt.title("Flowers Blooming by Day")
plt.xlabel("Day of the Year")
plt.ylabel("Flowers Count")

plt.tight_layout()

plt.show()

#Based on these histograms: I would propose that they should go between days 250 and 300 (Fall). The flight’s histogram shows fewer flights between these days, while the flowers histogram indicates there should be a number of flowers in bloom during this time.
```
